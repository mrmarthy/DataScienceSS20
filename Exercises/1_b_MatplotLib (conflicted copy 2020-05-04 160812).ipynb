## Exercise 1
Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: 

* X label name = Month Number
* Y label name = Total profit


```python
fig = plt.figure()
ax = plt.axes()

x = data.month_number
ax.plot(x, data.total_profit);
```

## Exercise 2
Get Total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

* Line Style dotted and Line color should be red
* Show legend at the lower right location.
* X label name = Month Number
* Y label name = Sold units number
* Add a circle marker.
* Line marker color as read
* Line width should be 3


```python
fig = plt.figure()
ax = plt.axes()

x = data.month_number
ax.plot(x, data.total_profit, ':g', label="Total Profit", marker='o', markerfacecolor="red", linewidth=3);

plt.xlabel('Month Number')
plt.ylabel('Sold units number')

ax.legend(loc='lower right');
```

## Exercise 3
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).

```python
data.moisturizer

fig = plt.figure()
ax = plt.axes()

x = data.month_number

ax.plot(x, data.moisturizer, label="moisturizer");
ax.plot(x, data.bathingsoap, label="bathingsoap");
ax.plot(x, data.shampoo, label="shampoo");
ax.plot(x, data.toothpaste, label="toothpaste");
ax.plot(x, data.facewash, label="facewash");
ax.plot(x, data.facecream, label="facecream");


plt.xlabel('Month Number')
plt.ylabel('Sold units number')

ax.legend(loc="upper center")
```

## Exercise 4
Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot.

```python
plt.scatter(data.month_number, data.toothpaste)
plt.xlabel('Month')
plt.ylabel('Sold units')
plt.title('scatter plot')
plt.grid()
plt.show()
```

## Exercise 5
Read face cream and facewash product sales data and show it using the bar chart
Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
plt.bar(data.month_number, data.facecream, width=-0.3, align='edge', label="Sold facecream Units")
plt.bar(data.month_number, data.facewash, width=0.3, align='edge', label="Sold facewash Units")
plt.xlabel('Month')
plt.ylabel('Sold units')
plt.title('Bar plot')
plt.legend()
plt.show()
```

## Exercise 6
Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

***Note:*** use ``data[item].sum()`` to compute sums over columns 

```python
data['toothpaste'].sum()
```

```python
labels = ['facecream', 'fasewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer'];
sums   = [data ['facecream'].sum(), data ['facewash'].sum(), data ['toothpaste'].sum(), 
         data ['bathingsoap'].sum(), data ['shampoo'].sum(), data ['moisturizer'].sum()];
plt.axis("equal");
plt.pie(sums, labels=labels, autopct='%1.1f%%')
```
