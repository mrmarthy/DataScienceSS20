````python
#setup env
import time
import warnings
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import cluster, datasets, mixture
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
````

````python
#helper calls
n_samples = 1500
random_state = 170
X, y = datasets.make_blobs(n_samples=n_samples, random_state=random_state)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)

#generating Data Sets A-F
A = datasets.make_circles(n_samples=n_samples, factor=.5,noise=.05)[0]
B = datasets.make_moons(n_samples=n_samples, noise=.05)[0]
C = datasets.make_blobs(n_samples=n_samples, random_state=8)[0]
D = np.random.rand(n_samples, 2)
E = (X_aniso, y)[0]
F = datasets.make_blobs(n_samples=n_samples,
                             cluster_std=[1.0, 2.5, 0.5],
                             random_state=random_state)[0]
````

Plot all raw data sets A-F in one figure.

## Exercise 2
Perform ```K-Means``` clustering on all data sets: https://scikit-learn.org/stable/modules/clustering.html#k-means
* 2.1 plot all results
* 2.2 manually try to find the best $k$ for each data set (by visual evaluation)
* 2.3 Use the ```K-Means++``` initialization 

````python


AKMeans = KMeans(n_clusters=2, init='k-means++')
BKMeans = KMeans(n_clusters=2, init='k-means++')
CKMeans = KMeans(n_clusters=3, init='k-means++')
DKMeans = KMeans(n_clusters=4, init='k-means++')
EKMeans = KMeans(n_clusters=2, init='k-means++')
FKMeans = KMeans(n_clusters=3, init='k-means++')

AKMeansfit = AKMeans.fit(A)
BKMeansfit = BKMeans.fit(B)
CKMeansfit = CKMeans.fit(C)
DKMeansfit = DKMeans.fit(D)
EKMeansfit = EKMeans.fit(E)
FKMeansfit = FKMeans.fit(F)


fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(6, figsize=(6,9))

ax1.scatter(A[:,0], A[:,1], c=AKMeansfit.labels_)
ax2.scatter(B[:,0], B[:,1], c=BKMeansfit.labels_)
ax3.scatter(C[:,0], C[:,1], c=CKMeansfit.labels_)
ax4.scatter(D[:,0], D[:,1], c=DKMeansfit.labels_)
ax5.scatter(E[:,0], E[:,1], c=EKMeansfit.labels_)
ax6.scatter(F[:,0], F[:,1], c=FKMeansfit.labels_)
````

## Exercise 3
Performs ```DBSCAN``` clustering on all data sets and plot all results: https://scikit-learn.org/stable/modules/clustering.html#dbscan

````python
ADB = DBSCAN(min_samples=2, eps=0.2)
BDB = DBSCAN(min_samples=4, eps=0.15)
CDB = DBSCAN(min_samples=2, eps=1)
DDB = DBSCAN(min_samples=2, eps=1)
EDB = DBSCAN(min_samples=4, eps=0.4)
FDB = DBSCAN(min_samples=2, eps=1)

ADBfit = ADB.fit(A)
BDBfit = BDB.fit(B)
CDBfit = CDB.fit(C)
DDBfit = DDB.fit(D)
EDBfit = EDB.fit(E)
FDBfit = FDB.fit(F)


fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(6, figsize=(6,9))

ax1.scatter(A[:,0], A[:,1], c=ADBfit.labels_)
ax2.scatter(B[:,0], B[:,1], c=BDBfit.labels_)
ax3.scatter(C[:,0], C[:,1], c=CDBfit.labels_)
ax4.scatter(D[:,0], D[:,1], c=DDBfit.labels_)
ax5.scatter(E[:,0], E[:,1], c=EDBfit.labels_)
ax6.scatter(F[:,0], F[:,1], c=FDBfit.labels_)
````
